```python
df_arrivals = tables[0] # I decided that my website shows flights for the current day.
df_arrivals
```

```python
df_arrivals = df_arrivals.drop(columns=["Unnamed: 4"]) # I deleted the column with flight details. 
df_arrivals
```

```python
%pip install --quiet deep-translator # The information is retrieved from a Polish website. 
from deep_translator import GoogleTranslator # I decided to translate it to English. 
```

```python
# Function to translate text, courtesy of ChatGPT.
def translate_text(text):
    try:
        return GoogleTranslator(source='pl', target='en').translate(text)
    except Exception as e:
        print(f"Error translating '{text}': {e}")
        return text

df_arrivals = df_arrivals.applymap(translate_text)
```

```python
# The function above translates only the values inside the dataframe, so for column names I did it manually.
df_arrivals = df_arrivals.rename(columns={"RozkÅ.": "Time",
               "Kierunek": "Direction",
               "Numer lotu": "Flight Number"
               })

df_arrivals["Arrival or Departure"] = "Arrival" # This dataframe contains arrivals only. 
# But since I'm planning to combine it with departures I created a separate column to clarify which flights are which. 

df_arrivals.head()
```

```python
# Now, I also want to account for the departures. 
df = pd.read_html("https://www.krakowairport.pl/pl/pasazer/loty/odloty")
df
```

```python
tables = pd.read_html("https://www.krakowairport.pl/pl/pasazer/loty/przyloty")
len(tables) # Again checking the website structure. 
```

```python
# Follow the same steps for the departures. 
df_departures = tables[0] 
df_departures = df_departures.drop(columns=["Unnamed: 4"])
df_departures["Arrival or Departure"] = "Departure"
df_departures
```

```python
# Translation
def translate_text(text):
    try:
        return GoogleTranslator(source='pl', target='en').translate(text)
    except Exception as e:
        print(f"Error translating '{text}': {e}")
        return text

df_departures = df_departures.applymap(translate_text)

df_departures = df_departures.rename(columns={"RozkÅ.": "Time",
               "Kierunek": "Direction",
               "Numer lotu": "Flight Number"
               })

df_departures.head()
```

```python
flights_krk = pd.concat([df_arrivals, df_departures])
flights_krk
```

```python
# Because I want to color code the flights that are cancelled or delayed, I created a separate column. 
# That way I can color code it in Datawrapper. 

def classify_flight(status_text):
    if isinstance(status_text, str):
        status_text = status_text.lower()
        if "cancel" in status_text:
            return "Cancelled"
        if "delay" in status_text:
            return "Delayed"
    return ""  

flights_krk["Delayed/ Cancelled"] = flights_krk["Status"].apply(classify_flight)
flights_krk
```

```python
flights_krk.to_csv("flights_krk.csv", index=False)
```
